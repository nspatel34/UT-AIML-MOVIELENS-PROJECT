### 8. Visualize how popularity of genres has changed over the years - 10 marks

Note that you need to use the **percent of number of releases in a year** as a parameter of popularity of a genre

Hint 1: You need to reach to a data frame where the release year is the index and the genre is the column names (one cell shows the number of release in a year in one genre) or vice versa. (Drop unnecessary column if there are any)

Hint 2: Find the total number of movies release in a year(use `sum(axis=1)` store that value in a new column as 'total'). Now divide the value of each genre in that year by total to get percentage number of release in a particular year.
`(df.div(df['total'], axis= 0) * 100)`

Once that is achieved, you can either use univariate plots or can use the heatmap to visualise all the changes over the years 
in one go. 

Hint 3: Use groupby on the relevant column and use sum() on the same to find out the number of releases in a year/genre.  

```python
# dropped the 'movie id' and 'movie title' and 'release date' to leave just 'release year'
df1aaa = df1.drop(['movie id', 'movie title', 'release date'], axis=1)
```

```python
# set index to 'release year'
df1aaa = df1aaa.set_index('release year')
```

```python
# sum of the columns and a new 'total' column for totals for each movie
df1aaa['total'] = df1aaa.sum(axis=1)
df1aaa
```

```python
# create a new df with dropped 'unknown column from df1aaa'
df1x = df1aaa.drop(['unknown'], axis=1)
```

```python
# group rows by release year and add up totals in each genre per year
df1xx = df1x.groupby('release year').agg(sum)
```

```python
#divide the value of each genre in that year by total to get percentage number of release in a particular year
df1aaaa = (df1xx.div(df1xx['total'], axis= 0) * 100)
df1aaaa
```

```python
# dropped 'total' so it doesn't mess up the heat map
df1xxx = df1aaaa.drop(['total'], axis=1)
```

```python
# heatmap how popularity of each genre has changed over the years
plt.figure(figsize=(10,15))
sns.heatmap(df1xxx, cmap="plasma")                                  
```

### 9. Find the top 25 movies according to average ratings such that each movie has number of ratings more than 100 - 10 marks

Hints : 

1. Find the count of ratings and average ratings for every movie.
2. Slice the movies which have ratings more than 100.
3. Sort values according to average rating such that movie which highest rating is on top.
4. Select top 25 movies.
5. You will have to use the .merge() function to get the movie titles.

Note: This question will need you to research about groupby and apply your findings. You can find more on groupby on https://realpython.com/pandas-groupby/.

```python
# Read the data
df1 = pd.read_csv('movie.csv')
df2 = pd.read_csv('rating.csv')
df3 = pd.read_csv('user.csv')
```

```python
dfm1 = df2.merge(df1, on= 'movie id', how='left')
dfm1
```

```python
dfm2= dfm1.set_index('movie title')
dfm2
```

```python
# find the count of ratings and average ratings for every movie
dfm3 = dfm2.drop(['timestamp','release date','unknown','movie id'], axis=1)
dfm3
```

```python
#Average Ratings
AR = pd.DataFrame(dfm2.groupby('movie title')['rating'].mean())
AR
```

```python
# Average Ratings and total number of ratings count per movie 
AR['Total_Ratings'] = pd.DataFrame(dfm3.groupby('movie title')['rating'].count())
AR
```

```python
# List all the movie with >100 ratings
ARx = AR[AR.Total_Ratings > 100]
ARx
```

```python
# the top 25 movies
ARx.sort_values(by='rating', ascending=False).head(25)
```
